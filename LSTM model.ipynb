```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
import pandas_datareader as data
```

```python
start = '2010-01-01'
end = '2019-12-31'

df = data.DataReader ('AAPL','yahoo',start,end)

df.head()
```

```python
df = df.reset_index()
df.head()
```

```python
df = df.drop(['Date','Adj Close'],axis = 1)
df
```

```python
# Simple graph to see the plot for close rates 
plt.plot(df.Close)
```

```python
#Now we are finding the moving avarage of 100 days 

ma100 = df.Close.rolling(100).mean()
ma100

# we have use the rolling function for getting the 100 days of avarage 
```

```python
plt.figure(figsize = (12,6)) 

plt.plot(df.Close)
plt.plot(ma100,'r') # r = red color 
```

```python
# now we just create 200 days of moving avarage 

ma200 = df.Close.rolling(200).mean()

plt.figure(figsize = (12,6)) 

plt.plot(df.Close)
plt.plot(ma100,'r')
plt.plot(ma200,'y') # r = red color 
```
